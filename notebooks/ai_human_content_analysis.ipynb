```python
import pandas as pd

#Data Cleaning

df = pd.read_csv('/Users/mac/Desktop/Mission Inferno/Data Analysis Project/data/dataframe.csv')

# Dataset dimension and basics info
df.shape
df.info()
```

```python
# DATA CLEANING

# Handling null values
df["flesch_reading_ease"] = df["flesch_reading_ease"].fillna(df["flesch_reading_ease"].median())
df["gunning_fog_index"] = df["gunning_fog_index"].fillna(df["gunning_fog_index"].median())
df["passive_voice_ratio"] = df["passive_voice_ratio"].fillna(df["passive_voice_ratio"].median())
df["sentiment_score"] = df["sentiment_score"].fillna(df["sentiment_score"].median())

# Duplicate records check
df = df.drop_duplicates()
df

```

```python
# DATA CONSISTENCY CHECKS

df_verify = df[["text_content", "word_count", "character_count", "sentence_count", "avg_sentence_length", "avg_word_length"]].copy()
df_verify["verify_avg_sentence_length"] = (df_verify["word_count"] / df_verify["sentence_count"]).round(2)
df_verify["verify_avg_word_length"] = (df_verify["character_count"] / df_verify["word_count"]).round(2)
df_verify["check_avg_sentence_length"] = df_verify["verify_avg_sentence_length"] == df_verify["avg_sentence_length"]
df_verify["check_avg_word_length"] = df_verify["verify_avg_word_length"] == df_verify["avg_word_length"]
df_verify
```

```python
# HANDLING INCONSISTENT DATA

df["avg_word_length"] = (df["character_count"] / df["word_count"]).round(2)
df["avg_sentence_length"] = (df["word_count"] / df["sentence_count"]).round(2)

df.to_csv('/Users/mac/Desktop/Mission Inferno/Data Analysis Project/data/exports/cleaned_dataframe.csv')

df
```

```python
# EXPLORATORY DATA ANALYSIS
```

```python
# Target Variable Analysis

ai_content = df[df["label"] == 1]
human_content = df[df["label"] == 0]

ai_content_count = ai_content["text_content"].count()
human_content_count = human_content["text_content"].count()

class_balance = {
    "Author": ["AI Generated", "Human Generated"],
    "Number of content": [ai_content_count, human_content_count],
    "Percentage": [((ai_content_count / (human_content_count + ai_content_count)) * 100).round(2), ((human_content_count / (human_content_count + ai_content_count)) * 100).round(2)]
}

class_balance = pd.DataFrame(class_balance)
class_balance.to_csv('/Users/mac/Desktop/Mission Inferno/Data Analysis Project/data/exports/class_balance.csv')
class_balance
```

```python
# Content Type Analysis

content_analysis = df.groupby(['content_type', 'label']).size().unstack(fill_value=0)
content_analysis.columns = ['Human Generated', 'AI Generated']
content_analysis["Total Count"] = content_analysis["Human Generated"] + content_analysis["AI Generated"]
content_analysis["Chance of AI Generated"] = ((content_analysis["AI Generated"] / content_analysis["Total Count"]) * 100).round(2)
content_analysis["Chance of Human Generated"] = ((content_analysis["Human Generated"] / content_analysis["Total Count"]) * 100).round(2)

content_analysis.to_csv('/Users/mac/Desktop/Mission Inferno/Data Analysis Project/data/exports/content_analysis.csv')
content_analysis
```

```python
# Text Length Characteristics

# Word Count Analysis

word_count_analysis = df.groupby(["content_type", "label"])["word_count"].agg(["median", "mean", "std"]).unstack()
word_count_analysis.columns = ["Human Generated Median", "AI Generated Median", "Mean Human Generated", "Mean AI Generated", "Std Dev Human Generated", "Std Dev AI Generated"]

word_count_analysis.to_csv('/Users/mac/Desktop/Mission Inferno/Data Analysis Project/data/exports/word_count_analysis.csv')
word_count_analysis
```

```python
# Character Count Analysis

character_count_analysis = df.groupby(["content_type", "label"])["character_count"].agg(["median", "mean", "std"]).unstack()
character_count_analysis.columns = ["Human Generated Median", "AI Generated Median", "Mean Human Generated", "Mean AI Generated", "Std Dev Human Generated", "Std Dev AI Generated"]

character_count_analysis.to_csv('/Users/mac/Desktop/Mission Inferno/Data Analysis Project/data/exports/character_count_analysis.csv')
character_count_analysis
```

```python
# Sentence Count Analysis

sentence_count_analysis = df.groupby(["content_type", "label"])["sentence_count"].agg(["median", "mean", "std"]).unstack()
sentence_count_analysis.columns = ["Human Generated Median", "AI Generated Median", "Mean Human Generated", "Mean AI Generated", "Std Dev Human Generated", "Std Dev AI Generated"]

sentence_count_analysis.to_csv('/Users/mac/Desktop/Mission Inferno/Data Analysis Project/data/exports/sentence_count_analysis.csv')
sentence_count_analysis
```
